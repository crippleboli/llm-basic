```python
from peft import LoraConfig
```

```python
# 1、构造LoRAConfig实例对象
lora_config = LoraConfig(
    r=8,
    lora_alpha=8,
    lora_dropout=0.05,
    bias="none",
    target_modules=["q_proj","v_proj"],
    task_type="CAUSAL_LM"
)
```

```python
# 2、通过get_peft_model获取peftmodel
from transformers import AutoModelForCausalLM
model = AutoModelForCausalLM.from_pretrained("model/qwen3-0.6B")    # 原始模型
```

```python
from peft import get_peft_model
peft_model = get_peft_model(model,lora_config)  # 冻结参数 植入旁路
```

```python
# 3. 查看原本的基座模型参数量（、
print(model.num_parameters())
```

```python
# 4. 查看 PEFT 模型
peft_model.print_trainable_parameters()
```

```python
# 7. 查看当前模型的结构：新增了A,B可训练模块
for name, module in peft_model.named_modules():
    print(name,module)
```
